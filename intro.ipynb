```python
aapl.shift(periods=1)
```

```python
aapl['lag1'] = aapl.shift(periods=1)
aapl
```

```python
aapl['diff'] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl
```

```python
aapl['% change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
aapl.AAPL.resample('M').last()
```

```python
aapl.AAPL.resample('BM').last().pct_change(periods=1).mul(100)
```

```python
aapl
del aapl['lag1']
del aapl['% change']
del aapl['diff']
```

```python
aapl
```

```python
ret = aapl.pct_change().dropna()
ret
```

```python
ret.plot(kind='hist', figsize=(8,5), fontsize=8, bins=100)
plt.show()
```

```python
daily_mean_return = ret.mean()
print(f'daily return:{daily_mean_return}')

var_daily = ret.var()
print(f'daily variance:{var_daily}')

daily_std = np.sqrt(var_daily)
print(f'daily standard deviation:{daily_std}')
```

```python
annual_mean_return = daily_mean_return * 252
annual_mean_return
```

```python
annual_var_return = var_daily * 252
annual_var_return
```

```python
annual_std_return = np.sqrt(annual_var_return)
annual_std_return
```

### Risk / Reward stocks

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe()
```

```python
ret.describe().T
```

```python
summary = ret.describe().T.loc[:, ['mean','std']]
summary
```

```python
summary['mean'] = summary['mean'] * 252
summary['std'] = summary['std'] * np.sqrt(252)
summary
```

```python
summary.plot(kind='scatter', x='std', y='mean', figsize=(10,8), s=30, fontsize=8)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, 'std'] + 0.005, summary.loc[i, 'mean'] + 0.005), size=10)
plt.xlabel('Annual risk (std)', fontsize=10)
plt.ylabel('Annual return', fontsize=10)
plt.title('Risk/Return', fontsize=10)
plt.show()
```

### Simple moving average

```python
SPY = yf.download('SPY')
```

```python
spy = SPY.Close.to_frame()
```

```python
spy
```

```python
spy.plot(figsize=(8,5), fontsize=(10))
plt.legend(loc='upper left', fontsize=10)
plt.show()
```

```python
spy_roll = spy.rolling(window=10)
```

```python
spy_roll = spy_roll.mean()
```

```python
spy_roll.head(15) # nan porque el moving average necesita dias anteriores
```

```python
spy.rolling(window=10).median()
```

```python
spy.rolling(window=10, min_periods=5).max()
```

```python
spy.rolling(window=10, min_periods=5).min().head(15)
```

```python
spy['SMA50'] = spy.rolling(window=50, min_periods=50).mean()
```

```python
spy.head(52)
```

```python
spy.plot(figsize=(8,5), fontsize=10)
plt.legend(loc='upper left', fontsize=10)
plt.show()
```
